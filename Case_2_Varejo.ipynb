```python
import pandas as pd
import seaborn as sns

vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
vendas.info() # informações detalhadas do dataset
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique() # quantidade de compras por canal
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP","Aplicativo") # alteração de nome da coluna
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_") # alteração de nome da coluna
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas[vendas["estado"].isnull()] # apresentação de dados nulos
```

```python
vendas_sem_nulos = vendas.dropna() # remoção de dados nulos
vendas_sem_nulos.info()
```

```python
vendas["estado"].fillna("MS") # substituindo os valores nulos por um estado
vendas.info()
```

```python
media_preco = vendas["Preço"].mean() # média do preço
vendas["Preço"].fillna(media_preco, inplace = True) # substituindo o valor do preço pela média
vendas.info()
```

```python
mediana_preco = vendas["Preço"].median() # mediana do preço
mediana_preco
#
```

```python
vendas[["Preço","Preço_com_frete"]].describe() # descrição estatística
```

```python
vendas_erro = vendas.query("Preço > Preço_com_frete") # verificação de erro no sistema
vendas_certo = vendas.query("Preço < Preço_com_frete")
vendas_certo.head()
```

```python
vendas_certo.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe() # descrição estatística após o filtro de algumas categorias
```

```python
# contagem de departamentos mais vendidos.

vendas_certo.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# média de preço por departamento

round(vendas_certo.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values( ascending=False).reset_index(),2)

#round determina a quantidade de números após o ponto
```

```python
# qtd de vendas por data

vendas_certo.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# criando coluna para extrair o mês do DataSet

vendas_certo["mes"] = vendas_certo["Data"].dt.month
vendas_certo.head()
```

```python
# qtd de vendas por Mês

vendas_certo.groupby("mes").idcompra.nunique().sort_values(ascending=False).reset_index()
```
