# Case Varejo

**KPIs solicitadas**:

Departamentos mais vendidos;

Média de Preço e Frete por Departamento;

Vendas por Mês;

Média de Renda por Canal;

Média de Idade por Bandeira.

**Premissas**

Erro no Sistema para estado Civil;

Validador de Preço x Frete.


Importação e visualização da planilha.

```python
import pandas as pd
import seaborn as sns

vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
vendas.info() # informações detalhadas do dataset
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique() # quantidade de compras por canal
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP","Aplicativo") # alteração de nome da coluna
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_") # alteração de nome da coluna
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas[vendas["estado"].isnull()] # apresentação de dados nulos
```

```python
vendas_sem_nulos = vendas.dropna() # remoção de dados nulos
vendas_sem_nulos.info()
```

```python
vendas["estado"].fillna("MS") # substituindo os valores nulos por um estado
vendas.info()
```

```python
media_preco = vendas["Preço"].mean() # média do preço
vendas["Preço"].fillna(media_preco, inplace = True) # substituindo o valor do preço pela média
vendas.info()
```

```python
mediana_preco = vendas["Preço"].median() # mediana do preço
mediana_preco
#
```

```python
vendas[["Preço","Preço_com_frete"]].describe() # descrição estatística
```

```python
vendas_erro = vendas.query("Preço > Preço_com_frete") # verificação de erro no sistema
vendas_certo = vendas.query("Preço < Preço_com_frete")
vendas_certo.head()
```

```python
vendas_certo.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe() # descrição estatística após o filtro de algumas categorias
```

```python
# contagem de departamentos mais vendidos.

vendas_certo.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# média de preço por departamento

round(vendas_certo.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values( ascending=False).reset_index(),2)

#round determina a quantidade de números após o ponto
```

```python
# qtd de vendas por data

vendas_certo.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# criando coluna para extrair o mês do DataSet

vendas_certo["mes"] = vendas_certo["Data"].dt.month
vendas_certo.head()
```

```python
# qtd de vendas por Mês

vendas_certo.groupby("mes").idcompra.nunique().sort_values(ascending=False).reset_index()
```

# Informações sobre Cliente

Utilizando a segunda base.

```python
import pandas as pd
import seaborn as sns

clientes = pd.read_excel('cliente_varejo.xlsx')
clientes.head()
```

```python
# devido as colunas iguais entre as bases de Varejo.xlsx e clientes

clientes[["idade", "renda"]].describe()
```

```python
clientes.info() # validação de dados nulos
```

```python
# Transformando a coluna de renda para float

clientes = clientes.astype({"renda":"float"})
clientes.info()
```

```python
# relacionando as duas bases através de um dado, tendo os dados das tabelas na mesma
vendas_cliente = vendas_certo.merge(clientes, how = "left", on = "cliente_Log")
vendas_cliente.head()
```

```python
# Média de renda por cada departamento

agg_idcanalvenda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").reset_index(),2)
agg_idcanalvenda
```

```python
# Média de idade por bandeira
agg_idade_bandeira= round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending= False).reset_index(),2)
agg_idade_bandeira
```

# Visualização em Gráfico das informações

```python
# Média de idade por bandeira

import matplotlib.pyplot as plt # importação da biblioteca

fig = plt.figure(figsize=(10,5))

plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color = "green")
plt.title("Média de Idade por Bandeira")
plt.xlabel("Bandeira")
plt.ylabel("Média de Idade")
```

```python
fig = plt.figure(figsize=(10,5))

plt.bar(agg_idcanalvenda["idcanalvenda"], agg_idcanalvenda["renda"], color = "blue")
plt.xlabel("Canal")
plt.ylabel("Média de Renda")
plt.title("Média de Renda por Canal", loc="center", fontsize = 18, fontweight="bold", color ="black");
```

```python
# representação de mudança de valores ao longo do tempo

venda_por_data = vendas_certo.groupby("Data").idcompra.nunique().reset_index()
venda_por_data

```

```python
fig = plt.figure(figsize=(20,12))

plt.xlabel("Data da Venda")
plt.ylabel("Quantidade de Vendas")
x = venda_por_data["Data"]
y = venda_por_data["idcompra"]

plt.plot(x, y, color="blue", linewidth=2)
plt.title("Venda por data", loc="center", fontsize=18, fontweight="bold", color="black")
plt.show()
```

```python
# média de preço por departamento
agg_departamento_preco = round(vendas_certo.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values( ascending=False).reset_index(),2)
agg_departamento_preco
```

```python
fig = plt.figure(figsize=(30,20))

plt.bar(agg_departamento_preco["Nome_Departamento"], agg_departamento_preco["Preço_com_frete"], color = "blue")
plt.xlabel("Departamento")
plt.ylabel("Média de Preço")
plt.title("Preço Médio por Departamento", loc="center", fontsize = 18, fontweight="bold", color ="black");
```
